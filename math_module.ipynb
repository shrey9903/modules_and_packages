```python


import math



def Fect():
    a=int(input('Enter number to find factorial: '))
    print('Fectorial:',math.factorial(a))


def Com_interest():
    a=float(input('Enter the amount:'))
    b=float(input('Enter interest rate(%):')) / 100
    c=float(input('Enter time in year:'))
    n=1
    A=a*(1+ b/n)**(n*c)
    print('Compound interest:',A)
    return

def Trig_calcu():
    angle= float(input("Enter angle in degrees: "))
    angle2 = math.radians(angle)
    print("Sin =", math.sin(angle2))
    print("Cos =", math.cos(angle2))
    print("Tan =", math.tan(angle2))
    return

def Area_of_space():
    while True:
        print("1. Area of Circle")
        print("2. Area of Rectangle")
        print("3. Area of Square")
        print("4. Area of Triangle")

        option=int(input('Enter youor option: '))
        match option:
            case 1:
                r = float(input("Enter radius: "))
                print("Area of Circle =", math.pi * r * r)
            case 2:
                l = float(input("Enter length: "))
                w = float(input("Enter width: "))
                print("Area of Rectangle =", l * w)
            case 3:
                s = float(input("Enter side: "))
                print("Area of Square =", s * s)
            case 4:
                b = float(input("Enter base: "))
                h = float(input("Enter height: "))
                print("Area of Triangle =", 0.5 * b * h)
            case _:
                print("Enter valid choice")
        return
def math_opration():
    while True:
        print("\nMathemetica oprations:")
        print('\n1.Calculate Factorial  ')
        print('2.Solve compound interest')
        print('3.Trigonometric calculation')
        print('4.Area of Geometric Shapes')
        print('5.Back to main manue')


        print('======'*15)
        option=int(input('Enter your option: '))
        print('======'*15)

        match option:
            case 1:#Calculate Factorial
                Fect()
            case 2:#Solve compound interest
                Com_interest()
            case 3:#Trigonometric calculation
                Trig_calcu()
            case 4:#Area of Geometric Shapes
                Area_of_space()
            case 5:#main manue
                break
            case _:
                print("Enter valid choice")
    return
math_opration()
```
