```python
import os


print('File Opration:')



def Crt_file(name):
    try:
        with open(name,"a") as f:
            print('======'*15)
            return f
    except Exception:
        print('Enter valid Name')
def Wri_file(name,entry):
    try:
        with open(name,"a") as f:
            f.write(entry)
            print('======'*15)
        return f
    except Exception:
        print('Enter valid Name')
def Red_file(name):
    try:
        with open(name) as f:
            print(f.readline())
            print('======'*15)
        return
    except Exception:
        print('Enter valid Name')
def App_file(name,entry):
    try:
        with open(name,"a") as f:
                f.write(entry)
                print('======'*15)
                return f
    except Exception:
        print('Enter valid Name')

def F_opration():
    while True:
        print('1.Create a new file')
        print('2.Write to a file')
        print('3.Read from a file')
        print('4.Append to a file')
        print('5.Back to main manu')

        option=int(input('Enter your option'))

        match option:
            case 1:#Create a new file
                try:
                    a=input('Enter file name with .txt:')
                    Crt_file(a)
                except Exception:
                    print('Enter valid Name')

            case 2:#Write to a file
                try:
                    a=input('Enter file name with .txt: ')
                    b=input('Enter data to write: ')
                    Wri_file(a,b)
                except Exception:
                    print('Enter valid Name')
            case 3:#Read from a file
                try:
                    a=input('Enter file name: ')
                    Red_file(a)
                except Exception:
                    print('Enter valid Name')
            case 4:#Append to a fileSS
                try:
                    a=input('Enter file name: ')
                    b=input('Enter data to Append: ')
                    App_file(a,b)
                except Exception:
                    print('Enter valid Name')
            case 5:#Back to main manu
                break
            case _:
                print("Enter valid choice")
    return
F_opration()
```
