```python
import random
import secrets
import string
print('======'*15)
print('Rendom Data Generation:')
print('======'*15)

def Ren_num():
    print(random.randint(1,10))
    print('======'*15)
    return
def Ren_list():
    random_list = [random.randint(1, 100) for _ in range(10)]
    print(random_list)
    print('======'*15)
    return

def Ren_pass(length=8):
        if length < 4:
            raise ValueError("Length must be at least 4")

        lower = secrets.choice(string.ascii_lowercase)
        upper = secrets.choice(string.ascii_uppercase)
        digit = secrets.choice(string.digits)
        special = secrets.choice(string.punctuation)

        all_chars = string.ascii_letters + string.digits + string.punctuation
        remaining = ''.join(secrets.choice(all_chars) for _ in range(length - 4))
        password_list = list(lower + upper + digit + special + remaining)
        random.shuffle(password_list)

        return ''.join(password_list)


def Ren_otp():
    otp = random.randint(100000, 999999)
    print("Your OTP is:", otp)
    print('======'*15)
    return
def ren_module():
    while True:

        print('1. Generate Random Number ')
        print('2. Generate Random List ')
        print('3. Create Rendom Password')
        print('4. Generate Random OTP ')
        print('5. Back to main manue ')


        option=int(input('Enter your option'))

        match option:
            case 1:#Generate Random Number
                Ren_num()
            case 2:#Generate Random List
                Ren_list()
            case 3:#Create Rendom Password
                print("Strong Password:", Ren_pass())
                print('======'*15)
            case 4:#Generate Random OTP
                Ren_otp()
            case 5:#Back to main manue
                break
            case _:
                print("Enter valid choice")
        return
ren_module()
```
